### *BARPLOT*

color = "blue" asigna el azul a todas las columnas

palette= ["red", "blue"] permite asignar un color distinto a cada columna

```python
sns.barplot(x = "sex", y ="total_bill", data = tipsdf)
plt.show()
```

```python
tipsdf[tipsdf.sex == "Male"].total_bill
```

```python
Male = tipsdf[tipsdf.sex == "Male"].total_bill.mean().round(2)
```

```python
Female = tipsdf[tipsdf.sex == "Female"].total_bill.mean().round(2)
```

```python
sns.barplot(x = "sex", y ="total_bill", data = tipsdf)
plt.annotate(Male, (-0.2,12), size = 25, color = "white")
plt.annotate(Female, (0.8, 10), size = 25, color = "white")
plt.show()
```

## TABLAS DINÁMICAS

Arreglo de los datos en un arreglo matricial diferente

Dataset = flights

```python
flightdf = sns.load_dataset("flights")
flightdf.head()
```

```python
flightdf.year.unique()
```

```python
flights = flightdf.pivot_table( index = "month", columns= "year", values= "passengers")
flights
```

```python
sns.heatmap(flights, cmap = "Greens", linecolor= "white", linewidths= 0.3)
plt.show()
```

#### Dataset = Iris (Información de flores)

```python
irisdf = sns.load_dataset("iris")
```

```python
irisdf
```

```python
irisdf.species.unique()
```

```python
irisdf.info()
```

```python
irisdf.species.value_counts()
```

```python
sns.countplot( x= "species", data= irisdf)
plt.show()
```

```python
irisdf.groupby(["species"]).count() # cuenta el número de registros para cada especie
```

```python
irisdf.groupby(["species"]).mean()
```

```python
irisdf.groupby(["species"]).median()
```

```python
irisdf.groupby(["species"]).max()
```

```python
irisdf.groupby(["species"]).min()
```

```python
sns.pairplot(irisdf, hue= "species")
plt.show()
```

```python
ax = sns.PairGrid(irisdf, hue= "species")
ax.map_lower(sns.kdeplot)
ax.map_upper(sns.scatterplot)
ax.map_diag(plt.hist)
plt.show()
```
